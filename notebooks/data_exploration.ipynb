```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file_1 = '../data/raw/tmidata_000_cv_splits.csv'
csv_file_2 = '../data/raw/tmidata_000_cv_splits_img.csv'
csv_file_3 = '../data/raw/tmidata_100rnaseq_cv_splits.csv'
csv_file_4 = '../data/raw/tmidata_100rnaseq_cv_splits_img.csv'

df1 = pd.read_csv(csv_file_1)
df2 = pd.read_csv(csv_file_2)
df3 = pd.read_csv(csv_file_3)
df4 = pd.read_csv(csv_file_4)
```

```python
print("Shape of df1:", df1.shape)
print("Shape of df2:", df2.shape)
print("Shape of df3:", df3.shape)
print("Shape of df4:", df4.shape)
```

```python
print("\nColumn names and data types in df1:")
print(df1.dtypes)
```

```python
# Check for missing values
print("Missing values in df1:")
print(df1.isnull().sum())

# Generate summary statistics for numerical columns
print("\nSummary statistics for numerical columns in df1:")
print(df1.describe())

# Visualize the distribution of a numerical column using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df1, x='Survival months', bins=30, kde=True)
plt.title('Distribution of Survival months')
plt.show()

# Visualize relationships between columns using a scatter plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df2, x='Grade', y='Survival months')
plt.title('Survival Months by Grade')
plt.xlabel('Grade')
plt.ylabel('Survival Months')
plt.show()
```
